## Working with Images

By leveraging the power of some common libraries that you can install, such as PILLOW, Python gains the ability to work with and manipulate images for simple tasks.
Install library by :

pip install pillow

```python
from PIL import Image
```

```python
mac = Image.open('example.jpg')

mac.show() # or simply mac on jupyter
```

```python
mac.filename
```

```python
mac.format_description
```

```python
mac.size
```

```python
halfway = 1993/2
```

```python
x = halfway - 200
w = halfway + 200
```

```python
y = 800
h = 1257
```

```python
computer = mac.crop((x,y,w,h))
computer
```

```python
mac.paste(im=computer, box=(800,0))
```

```python
mac
```

```python
# RGBA - A for alpha transparency
red = Image.open('red_color.jpg')
```

```python
red.putalpha(128)
red
```

```python
blue = Image.open('blue_color.jpg')
```

```python
blue.putalpha(128)
blue
```

```python
blue.paste(im=red, box=(0,0), mask=red)
```
