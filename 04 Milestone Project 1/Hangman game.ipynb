```python
stages = ['''
  +---+
  |   |
      |
      |
      |
      |
=========
''',
'''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''',
'''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', 
'''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''',
'''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''',
'''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
'''     ]
```

```python
word_list = ['camel', 'alligator','chimpanzee','leopard']
selected_word = random.choice(word_list).lower()
display = ['_']

display*= len(selected_word)
print ('Hint: animal name')
#print (display)
```

```python
tries = []
game_on = True

while game_on:
    
    guess = input ('Enter the letter u wanna guess: ').lower()
    for position in range(len(selected_word)):
        letter = selected_word[position]
        if letter == guess:
            display[position] = letter
            #print (display)

    if guess not in selected_word:
        tries.append(guess)

    if len(tries) == 6:
        game_on = False
        print ('Dead, u Hanged the man')

    print(f"{' '.join(display)}")
    
    if '_' not in display:
        game_on = False
        print ('You have won')

    print (stages[len(tries)])
```
